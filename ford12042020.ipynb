```python
def get(tickers, startdate, enddate):
    def data(ticker):
        return(pdr.get_data_yahoo(ticker, start=startdate, end=enddate))
    datas = map(data, tickers)
    return(pd.concat(datas, keys=tickers, names=['Ticker', 'Date']))

tickers = ['F', 'GM']
all_data = get(tickers, datetime.datetime(2019, 6, 1), datetime.datetime(2020, 4,8))
print(all_data)
```

```python
daily_close_px = all_data[['Adj Close']].reset_index().pivot('Date', 'Ticker', 'Adj Close')
daily_pct_change = daily_close_px.pct_change()
daily_pct_change.hist(bins=50, sharex=True, figsize=(12,8))
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(daily_pct_change, diagonal='kde', alpha=0.1, figsize=(12,8))
plt.show()
```

```python
adj_close_px = ford['Adj Close']
moving_avg = adj_close_px.rolling(window=40).mean()
print(moving_avg[-10:])
```

```python
ford['42']= adj_close_px.rolling(window=40).mean()
ford['252'] = adj_close_px.rolling(window=252).mean()
ford[["Adj Close", "42", "252"]].plot()
plt.show()
```

```python
short_window = 40
long_window = 100

signals = pd.DataFrame(index=ford.index)
signals['signal'] = 0.0

signals['short_mavg'] = ford['Close'].rolling(window=short_window, min_periods=1, center=False).mean()
signals['long_mavg']  = ford['Close'].rolling(window=long_window, min_periods=1,  center=False).mean()

signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:]
                                           > signals['long_mavg'][short_window:], 1.0, 0.0)

signals['positions'] = signals['signal'].diff()

print(signals)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111, ylabel='Preturi in $')

ford['Close'].plot(ax=ax1, color='r', lw=2.)

signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

ax1.plot(signals.loc[signals.positions ==1.0].index,
        signals.short_mavg[signals.positions == 1.0],
        '^', markersize=10, color='m')

ax1.plot(signals.loc[signals.positions == -1.0].index,
        signals.short_mavg[signals.positions == -1.0],
        'v', markersize=10, color='k')

plt.show()
```
